***Types of Encoding***

*Nominal Encoding*
 Techniques : One hot encoding, One hot encoding with many categorical variables, Mean Encoding
*Ordinal Encoding*
 Techniques : Label Encoding, Target Guided Ordinal Encoding


**One hot Encoding**
No of unique lable  ==  No of colums added.
Dummy Variable Trap, always one column can detect if all column value is zero.
To need to avoid that the total new column is added ==> n-1

Disadvantage : Curse of Dimensionality


**Label Encoding**
give label to values respect to order.

**One hot Encoding With multiple Categories**
use only one hot encoding to top_10 categories.


**Target Guided Ordinal Categories**
feature and output
A 1
B 1
C 0
A 0
B 1
......

Find mean respect to categories
based on mean , assign rank to categories as labels

There is chance of data leakage , must use training data only to calculate mean.

**Mean Encoding**
feature and output
A 1
B 1
C 0
A 0
B 1
..
find mean each categorie
use that mean.

**Count / Frequency encoding**

multiple number of categories == high cardinality

Replace each label of the categorical variable by the count , this is the amount of times each label appears in the dataset.

simple to implement
does not increase curse of feature dimensionality

if some of the lables have the same coutn then they will ve replaced with same count and they loose some valuable information


```python
import pandas as pd
```

```python
df = pd.read_csv('data/mercedesbenz.csv',usecols=['X1'])
df
```

```python
data_dict = df.X1.value_counts().to_dict()
```

```python
data_dict
```

```python
df['X1'] = df['X1'].map(data_dict)
df.head()
```

**Probability Ratio Encoding**

Also called Odds Ratio Encoding.

Instead of:

One-Hot Encoding

Label Encoding

We replace each category with:


P(Not Survived | Category)
divide
P(Survived | Category)
	​
Why Use It?

Because it captures:

“How strongly this category is associated with survival?”

Instead of creating many dummy columns.


```python
import pandas as pd
```

```python
df = pd.read_csv('data/titanic.csv',usecols=['Cabin','Survived'])
df.head()
```

```python
df['Cabin'] = df['Cabin'].fillna('Missing')
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Cabin'] = df['Cabin'].astype(str).str[0]
df.head()
```

```python
df.Cabin.unique()
```

```python
prod_df = df.groupby(['Cabin'])['Survived'].mean()
prod_df
```

```python
prod_df = pd.DataFrame(prod_df)
prod_df
```

```python
prod_df['Died'] = 1 - prod_df['Survived']
prod_df.head()
```

```python
prod_df['Probability_ratio'] = prod_df['Survived']/prod_df['Died']
prod_df.head()
```

```python
prob_encode = prod_df['Probability_ratio'].to_dict()
prob_encode
```

```python
df['Cabin_encoded'] = df['Cabin'].map(prob_encode) 
df.head()
```
