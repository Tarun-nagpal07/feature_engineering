## working with dates

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
data.info()
```

```python
data['year'] = data['date'].dt.year
```

```python
data['month'] = data['date'].dt.month
```

```python
data['month_name'] = data['date'].dt.month_name()
```

```python
data['days'] = data['date'].dt.day
```

```python
data['day_of_week'] = data['date'].dt.day_of_week
```

```python
data['dow_name'] = data['date'].dt.day_name()
```

```python
data['isweekend'] = np.where(data['dow_name'].isin(['Sunday','Saturday']),1,0)
```

```python
data['weekofyear'] = data['date'].dt.weekday
```

```python
data['quarter'] = data['date'].dt.quarter
```

```python
data['semester'] = np.where(data['quarter'].isin([1,2]), 1, 2)
```

```python
data.head()
```

```python
import datetime

today = datetime.datetime.today()

today
```

```python
today - data['date']
```

```python
(today - data['date']).dt.days
```

```python
# Months passed

np.round((today -data['date']) / np.timedelta64(1, 'M'),0)
```
