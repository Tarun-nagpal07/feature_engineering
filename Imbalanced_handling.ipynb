# 🔷 What Is an Imbalanced Dataset?

When one class has **much more samples** than the other.

### Example:

| Class         | Count |
| ------------- | ----- |
| Not Fraud (0) | 9900  |
| Fraud (1)     | 100   |

Here:

* 99% = Majority class
* 1% = Minority class

---

# 🔥 Why Is It a Problem?

If model predicts:

> Always predict “Not Fraud”

Accuracy = 99% 😳

But model is useless.

So **accuracy becomes misleading**.

---

# 🔷 Step 1: Use Correct Evaluation Metrics

DO NOT rely on accuracy.

Use:

* Precision
* Recall
* F1-score
* ROC-AUC
* PR-AUC (very important for imbalance)
* Confusion Matrix

---

# 🔷 Step 2: Techniques to Handle Imbalance

We divide them into:

1️⃣ Data-Level Methods
2️⃣ Algorithm-Level Methods
3️⃣ Hybrid / Advanced Methods

---

# 🔶 1️⃣ Data-Level Methods (Resampling)

---

## 🔹 A. Undersampling (Reduce Majority Class)

Randomly remove samples from majority class.

Example:
9900 → reduce to 100

### Pros:

✔ Simple
✔ Faster training

### Cons:

❌ Lose information
❌ Risk of underfitting

Best for:

* Very large datasets

---

## 🔹 B. Oversampling (Increase Minority Class)

Duplicate minority samples.

100 → 9900

### Pros:

✔ No data loss

### Cons:

❌ Overfitting (duplicate rows)

---

## 🔹 C. SMOTE (Synthetic Minority Over-sampling Technique)

Instead of duplicating, create synthetic points.

It:

* Picks a minority sample
* Finds nearest minority neighbors
* Creates new synthetic sample between them

---

### Pros:

✔ Better than simple oversampling
✔ Reduces overfitting

### Cons:

❌ Can create noisy samples

Best for:

* Medium sized datasets
* Continuous features

---

## 🔹 D. ADASYN

Similar to SMOTE but focuses more on hard-to-learn minority samples.

Used when:

* Minority class is complex

---

# 🔶 2️⃣ Algorithm-Level Methods

Instead of changing data, change model behavior.

---

## 🔹 A. Class Weighting

Assign higher penalty to minority class mistakes.

Example:

```python
class_weight = {0:1, 1:10}
```

Now misclassifying class 1 is 10x costly.

---

### Used In:

* Logistic Regression
* SVM
* Neural Networks
* Random Forest
* XGBoost

---

### Pros:

✔ No data manipulation
✔ Clean solution

Best for:

* Large datasets

---

## 🔹 B. Cost-Sensitive Learning

Define custom cost matrix.

Example:

| Actual | Pred 0 | Pred 1 |
| ------ | ------ | ------ |
| 0      | 0      | 1      |
| 1      | 10     | 0      |

Used when:

* False negative is dangerous (medical, fraud)

---

# 🔶 3️⃣ Advanced Ensemble Techniques

---

## 🔹 A. Balanced Random Forest

Each tree:

* Uses balanced bootstrap sample

Good for:

* High imbalance problems

---

## 🔹 B. EasyEnsemble

Trains multiple models on different undersampled subsets.

---

## 🔹 C. XGBoost / LightGBM with scale_pos_weight

Set:

```python
scale_pos_weight = majority/minority
```

Very effective for:

* Large tabular data

---

# 🔷 4️⃣ Threshold Tuning

Instead of default 0.5 probability:

Adjust threshold.

Example:

```python
if prob > 0.2:
    predict 1
```

Improves recall.

---

# 🔷 5️⃣ Anomaly Detection Approach

When minority extremely rare (0.1%):

Use:

* Isolation Forest
* One-Class SVM
* Autoencoders

Treat minority as anomaly.

---

# 🔷 Summary Table

| Method           | Type            | Data Modified | Overfitting Risk |
| ---------------- | --------------- | ------------- | ---------------- |
| Undersampling    | Data            | Yes           | Low              |
| Oversampling     | Data            | Yes           | High             |
| SMOTE            | Data            | Yes           | Medium           |
| Class Weights    | Algorithm       | No            | Low              |
| Balanced RF      | Ensemble        | Partial       | Low              |
| Threshold Tuning | Post-processing | No            | Low              |

---

# 🔷 When to Use What?

| Scenario                | Best Approach              |
| ----------------------- | -------------------------- |
| Small dataset           | SMOTE                      |
| Huge dataset            | Class weights              |
| Fraud detection         | XGBoost + scale_pos_weight |
| Medical (FN costly)     | Cost-sensitive learning    |
| Extreme imbalance (<1%) | Anomaly detection          |

---

# 🔥 Real ML Pipeline Strategy

Best practice:

1. Use stratified split
2. Use SMOTE on training only
3. Use class weights
4. Tune threshold
5. Evaluate with PR-AUC

---

# 🚀 Important Concept

Never oversample before train-test split ❌
Always oversample only training data.

Otherwise → data leakage.
