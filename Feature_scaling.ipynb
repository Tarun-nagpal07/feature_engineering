some of ml algo must require feature scaling

many feature has different units and highly magnitude difference

1. Linear Regression
2. Kmeans
3. KNN
they both use Euclidean Distance, it must to scale the values.
etc

where not required:
1. Decision Tree
2. Random Forest 
3. Ensumble Techniques
